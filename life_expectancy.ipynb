```python
df = pd.read_csv('Life_expectancy.csv')
```

```python
# Create seperate dataframes for each country
```

```python
countries = df['Entity'].unique().tolist()
```

```python
df_country = []

for i in countries:
    df_country.append(
        df[
            df['Entity'] == i
        ]
    )
```

```python
# Plot life expectancy per country
```

```python
plt.style.use('seaborn-talk')

for j,i in enumerate(df_country):
    plt.figure(figsize=(20,10))
    plt.plot(i['Year'], i['Life expectancy'])
    plt.title('Life expectancy for ' + countries[j])
    plt.xlabel('Years')
    plt.ylabel('Average Life Expectancy')
    _ = plt.xticks(rotation=45)
    plt.show()
```
